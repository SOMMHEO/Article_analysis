### Naver article Crawler

2024.06.15 update

```python
from tqdm import tqdm
import urllib.request
from urllib.request import Request, urlopen
from urllib import request
import openpyxl
from bs4 import BeautifulSoup

def naver_crawling(start, end, keyword):   
    wb = openpyxl.Workbook()
    sheet = wb.active
    sheet.append(['title','agency','date'])

    for i in tqdm(range(18)):
        try:
            start_ori = start.replace('.', '')
            end_ori = end.replace('.', '')
            
            url = (
                    "https://search.naver.com/search.naver?where=news&sm=tab_pge&query="
                    + keyword
                    + "&sort=0&photo=0&field=0&pd=3&ds="
                    + start
                    + "&de="
                    + end
                    + "&cluster_rank=21&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:from"
                    + start_ori
                    + "to"
                    + end_ori
                    + ",a:all&start="
                    + str(i*10+1)
                )
            print(url)


            # HTTP 403 Error 해결
            request_site = Request(url, headers={"User-Agent": "Mozilla/5.0"})
            webpage = urlopen(request_site).read()

            soup = BeautifulSoup(webpage, 'html.parser')
            
            # 하나의 article box
            articles = soup.select("#main_pack > section > div.api_subject_bx > div.group_news > ul > li") # 확인 및 변경
            
            for article in articles:
                try:
                    #기사제목
                    a_tag = article.select_one('a.news_tit') #sp_nws1 > div > div > div.news_contents > a.news_tit
                    title = a_tag.text
                    #언론사
                    comp = article.select_one('a.info.press').text.split(' ')[0]
                    #기사날짜
                    date = article.select_one('span.info').text
                    sheet.append([title, comp, date])
                except Exception as e:
                    continue
                
        except Exception as e:
                    break
    
    root_name = '뉴스_'+start
    root = 'C:/'+root_name+'.xlsx' # 경로 설정

    wb.save(root)
```

```python
# 사용방법
naver_crawling('2021.01.01','2021.01.31','5G')
```
